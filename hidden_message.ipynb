```python
import numpy as np
from scipy import misc
from skimage import filters
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = misc.imread('dog.jpeg')
msg = misc.imread('message.jpeg', mode='L')
thresh = filters.threshold_otsu(msg)
```

```python
plt.imshow(img);
```

```python
plt.hist(img.ravel())
```

```python
plt.imshow(msg)
```

```python
img.shape, msg.shape
```

```python
fft = np.fft.fft2(img)

fft[:,:,0][msg < thresh] = 30

plt.imshow(20*np.log10(abs(fft[:,:,0])));
```

```python
img2 = np.abs(np.fft.ifft2(fft)).astype(np.uint8)
```

```python
plt.hist(img2.ravel())
```
